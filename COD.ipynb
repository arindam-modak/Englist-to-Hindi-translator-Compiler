## Importing Required Files - NLTK

```python
import nltk 
#nltk.download()
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize, sent_tokenize 
from nltk.stem import *
```

 ## Gender Checking Function

```python
def gender_check(a):
    if(a[1] == "PRP"):
        if(a[0] == "He" or a[0] == "he" or a[0] == "I"):
            return 0
        elif(a[0] == "She" or a[0] == "she"):
            return 1
        
    if(a[1] != "NNP"):
        return 2
    
    if(a[0][-1] in "aeiouy"):
        return 1
    return 0
```

## Importing Dataset 

```python
#Storing the dictionary
Words_dict = {}

stemmer = WordNetLemmatizer()
sentences = open('Sentence2.txt','r')
sentences = sentences.readlines()

array = []

sentences = [sentence.rstrip() for sentence in sentences]
for sentence in sentences:
    wordsList = nltk.word_tokenize(sentence)
    wordsList = [word for word in wordsList]
    tagged = nltk.pos_tag(wordsList) 
    print(tagged)
    
    var = gender_check(tagged[0])
    
    #Classify the gender along the label
    
    for tup in tagged:
        if(tup in array):
            continue

        else:
            if((var == 1 and tup[1][0] == 'V') or (var == 1 and tup[1][0] == 'J')):
                array.append((tup[0],tup[1],'F'))
                array.append((tup[0],tup[1],'M'))
                array.append((tup[0],tup[1],'O'))
                
            elif ((var == 0 and tup[1][0] == 'V') or (var == 0 and tup[1][0] == 'J')):
                array.append((tup[0],tup[1],'M'))
                array.append((tup[0],tup[1],'F'))
                array.append((tup[0],tup[1],'O'))
                
            else:
                array.append((tup[0],tup[1],'O'))
                
#List of tuples
array
```

## Reading the words for translation from english to hindi

```python
#Reading the dictionary for translation

dictionary = open('dict2.txt','r',encoding = 'utf-8')
all_words = dictionary.readlines()

all_words = [x.rstrip() for x in all_words]

data = {}

for trans in all_words :
    temp = trans.split(" ")
    data[(temp[0],temp[1],temp[2])] = " ".join(temp[3:])

#Lexemes for translation.
data
```

## Converting input statement and matching the rule in Rules file

```python
wordsList = nltk.word_tokenize(statement)
wordsList = [word for word in wordsList]
tagged = nltk.pos_tag(wordsList) 

#Tagged contains the tagged word of the input statement

word_list = []
var = gender_check(tagged[0])
for tup in tagged:
    
    if((var == 1 and tup[1][0] == 'V') or (var == 1 and tup[1][0] == 'J')):
        word_list.append((tup[0],tup[1],'F'))
    elif ((var == 0 and tup[1][0] == 'V') or (var == 0 and tup[1][0] == 'J')):
        word_list.append((tup[0],tup[1],'M'))
    else :
        word_list.append((tup[0],tup[1],'O'))
        
#Breaking into lexemes
print(word_list)
        
#Mapping of Rules to Tuple
dummy_word_list = {}
for tup in tagged:
    if((var == 0 and tup[1][0] == 'V') or (var == 0 and tup[1][0] == 'J')):
        dummy_word_list[tup[1]] = (tup[0],'M')
    elif((var == 1 and tup[1][0] == 'V') or (var == 1 and tup[1][0] == 'J')):
        dummy_word_list[tup[1]] = (tup[0],'F')
    else:
        dummy_word_list[tup[1]] = (tup[0],'O')
        
word_str = ""

for wor in word_list:
    word_str = word_str + wor[1] + ','
    
word_str = word_str[:-1]
word_str
```
